```python
import pandas as pd
df = pd.read_csv("student_feedback.csv")
```

```python
df = df.drop(columns=["Unnamed: 0", "Student ID"])
```

```python
df.columns = [
    "Subject_Knowledge",
    "Concept_Clarity",
    "Use_of_Presentations",
    "Assignment_Difficulty",
    "Doubt_Solving",
    "Course_Structure",
    "Extra_Support",
    "Course_Recommendation"
]
```

```python
print("Updated Columns:\n", df.columns.tolist())
```

```python
print("Any Missing Values:\n", df.isnull().sum().sum())
```

```python
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
avg_ratings = df.mean().sort_values(ascending=False)

plt.figure(figsize=(10,5))
sns.barplot(x=avg_ratings.index, y=avg_ratings.values, palette="crest")
plt.xticks(rotation=45, ha='right')
plt.title("Average Student Ratings per Parameter", fontsize=14, weight='bold')
plt.ylabel("Average Rating (out of 10)")
plt.xlabel("")
plt.show()

print("Average Ratings:\n", avg_ratings)
```

```python
corr_matrix = df.corr()
print("Correlation Matrix:\n", corr_matrix.round(2))

plt.figure(figsize=(8,6))
sns.heatmap(corr_matrix, annot=True, cmap="YlGnBu", fmt=".2f")
plt.title("Correlation between Student Feedback Parameters", fontsize=14, weight='bold')
plt.show()
```

```python
avg_scores = df.mean().sort_values(ascending=False)

plt.figure(figsize=(8,5))
sns.barplot(x=avg_scores.values, y=avg_scores.index, palette="viridis")
plt.title("Average Student Feedback Scores", fontsize=14, weight='bold')
plt.xlabel("Average Rating (Out of 10)")
plt.ylabel("Feedback Categories")
plt.show()

print("Average Ratings:\n", avg_scores.round(2))
```
